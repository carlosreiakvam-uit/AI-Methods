```python
from sklearn.tree import DecisionTreeRegressor
from utils import *
import pandas as pd
import matplotlib.pyplot as plt
```

#### 2a
*Implement a function to read the file “data_2021.csv” and create a data set for a regression
tree. Use the sliding window approach (6 pts)*

```python
def read_data_2021():
    return pd.read_csv('data/data_2021.csv')
```

```python
window_size = 10
```

```python
pd_2021 = read_data_2021()
np_2021 = pd_2021['Demand'].to_numpy()
x, y = slide(np_2021, window_size)
```

#### 2b)
*Create a function to train a single regression tree on the dataset you just created (5 pts)*

```python
tree = DecisionTreeRegressor()
tree.fit(x, y)
```

#### 2c
*Using the same sliding window approach as in (2.a), start from 01.01.2021 and use the
regression tree to predict the next value. Save each prediction in a list. Use matplotlib to plot
the results alongside the original data.*

```python
def predict_2021(window_size):
    pred_2021 = [None for _ in range(window_size)]
    for window in x:
        pred = tree.predict([window])
        pred_2021.append(pred[0])
    return pred_2021
```

```python
predictions = predict_2021(window_size)
```

```python
ax1 = pd_2021.plot(y='Demand')
df_pred = pd.DataFrame(predictions, columns=['Prediction'])
df_pred.plot(grid=True, rot=90, ax=ax1)
plt.show()
```
