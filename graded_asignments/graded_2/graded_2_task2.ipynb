```python
from graded_asignments.graded_2.regressor import Regressor
from graded_asignments.graded_2.utils import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_data_2021():
    return pd.read_csv('data/data_2021.csv')
```

```python
window_size = 10
```

```python
pd_2021 = read_data_2021()
np_2021 = pd_2021['Demand'].to_numpy()
x, y = make_dataset(np_2021, window_size)
```

```python
model = Regressor()
tree = model.tree
tree.fit(x, y)
```

```python
def predict_2021(window_size):
    pred_2021 = [0 for _ in range(window_size)]
    for window in x:
        pred = tree.predict([window])
        pred_2021.append(pred[0])
    return pred_2021
```

```python
predictions = predict_2021(window_size)
```

```python
ax1 = pd_2021.plot(y='Demand')
df_pred = pd.DataFrame(predictions, columns=['Prediction'])
df_pred.plot(grid=True, rot=90, ax=ax1)
plt.show()
```
